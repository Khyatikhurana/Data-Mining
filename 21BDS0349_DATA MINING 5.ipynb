K-Medoid Clustering with Wine Dataset

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn_extra.cluster import KMedoids

# Load the Wine dataset
wine = load_wine()
X = wine.data  # Features

# Apply PCA to reduce dimensionality to 2D for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Apply K-Medoids clustering
kmedoids = KMedoids(n_clusters=3, random_state=42)
kmedoids.fit(X_reduced)

# Print the cluster labels and centers
print("Cluster Labels:")
print(kmedoids.labels_)
print("\nCluster Centers (in reduced space):")
print(kmedoids.cluster_centers_)

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=kmedoids.labels_, cmap='rainbow', s=50)
plt.scatter(kmedoids.cluster_centers_[:, 0], kmedoids.cluster_centers_[:, 1], 
            c='black', marker='x', s=200, label='Cluster Centers')
plt.title('K-Medoids Clustering (Wine Dataset)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.legend()
plt.show()
```

Agglomerative Clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch

# Load the Wine dataset
wine = load_wine()
X = wine.data  # Features

# Plot dendrogram for hierarchical clustering
plt.figure(figsize=(10, 7))
dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))
plt.title('Dendrogram for Agglomerative Clustering (Wine Dataset)')
plt.xlabel('Samples')
plt.ylabel('Euclidean distances')
plt.show()

# Apply Agglomerative clustering
agglom = AgglomerativeClustering(n_clusters=3)
y_agglom = agglom.fit_predict(X)

# Print the cluster labels
print("Cluster Labels for Agglomerative Clustering:")
print(y_agglom)

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_agglom, cmap='rainbow', s=50)
plt.title('Agglomerative Clustering (Wine Dataset)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

Divisive clustering

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load the Wine dataset
wine = load_wine()
X = wine.data  # Features

# Apply PCA to reduce dimensionality to 2D for visualization
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)

# Divisive clustering using recursive K-Means
def divisive_clustering(X, num_clusters):
    clusters = [X]  # Start with the whole dataset as one cluster
    cluster_labels = np.zeros(X.shape[0])  # Initialize cluster labels

    for _ in range(num_clusters - 1):
        largest_cluster_index = np.argmax([len(c) for c in clusters])
        to_split = clusters[largest_cluster_index]

        # Apply K-Means to split the largest cluster
        kmeans = KMeans(n_clusters=2, random_state=42)
        labels = kmeans.fit_predict(to_split)

        # Separate the largest cluster into two new clusters
        cluster_1 = to_split[labels == 0]
        cluster_2 = to_split[labels == 1]

        # Update clusters list
        clusters.pop(largest_cluster_index)
        clusters.append(cluster_1)
        clusters.append(cluster_2)

        # Update cluster labels
        cluster_labels[np.where(cluster_labels == largest_cluster_index)[0][labels == 1]] = len(clusters) - 1

    return cluster_labels

# Perform divisive clustering to split into 3 clusters
y_divisive = divisive_clustering(X, num_clusters=3)

# Print the cluster labels
print("Cluster Labels for Divisive Clustering:")
print(y_divisive)

# Plot the results
plt.figure(figsize=(8, 6))
plt.scatter(X_reduced[:, 0], X_reduced[:, 1], c=y_divisive, cmap='rainbow', s=50)
plt.title('Divisive Clustering (Wine Dataset)')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
