```python
print("Khyati Khurana - 21BDS0349")
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt

def page_rank(links, num_iterations=100, d=0.85):
    n = len(links)
    ranks = np.ones(n) / n
    M = np.zeros((n, n))

    for i, link in enumerate(links):
        for j in link:
            M[j][i] = 1 / len(link)

    for _ in range(num_iterations):
        ranks = (1 - d) / n + d * np.dot(M, ranks)
    
    return ranks

def visualize_pagerank(links, pagerank):
    G = nx.DiGraph()
    
    for i, link in enumerate(links):
        for j in link:
            G.add_edge(i, j)
    
    plt.figure(figsize=(8, 8))
    pos = nx.spring_layout(G)
    node_size = [5000 * rank for rank in pagerank]
    nx.draw(G, pos, with_labels=True, node_size=node_size, node_color=node_size, cmap=plt.cm.Blues, font_weight='bold')
    plt.title('PageRank Visualization')
    plt.show()

# Example link structure
links = [[1, 2], [2], [0], [0, 2]]
pagerank = page_rank(links)

# Print PageRank values
print("Page Ranks:", pagerank)

# Visualize the PageRank
visualize_pagerank(links, pagerank)
```

```python
print("Khyati Khurana - 21BDS0349")
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import seaborn as sns
import matplotlib.pyplot as plt

def document_similarity(doc1, doc2):
    vectorizer = CountVectorizer().fit_transform([doc1, doc2])
    vectors = vectorizer.toarray()
    
    # Compute the cosine similarity
    cosine_sim = cosine_similarity(vectors)
    
    # Print the cosine similarity between the two documents
    print(f"Cosine Similarity between Doc1 and Doc2: {cosine_sim[0, 1]:.4f}")
    
    return cosine_sim

def visualize_document_similarity(docs):
    # Compute the cosine similarity matrix
    cosine_sim = document_similarity(docs[0], docs[1])
    
    # Visualize the similarity matrix
    sns.heatmap(cosine_sim, annot=True, cmap='coolwarm', xticklabels=['Doc1', 'Doc2'], yticklabels=['Doc1', 'Doc2'])
    plt.title("Document Similarity Heatmap")
    plt.show()

# Example documents
doc1 = "The quick brown fox jumps over the lazy dog"
doc2 = "A fast brown fox leaps over the lazy hound"
docs = [doc1, doc2]

# Visualize and print similarity
visualize_document_similarity(docs)
```
